```python
# Load User Ids from file

user_ids = []
with open('data/random_users_million.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter='\n')
    for row in reader:
        user_ids.append(row[0])
```

```python
header_row = ['user_id_str', 'bio', 'location', 'name', 'screen_name', 'protected',
             'verified', 'followers_count', 'friends_count', 'statuses_count']

user_fields=["name", "username", "location", "protected", 
             "verified", "description", "public_metrics"]

with open('user_census.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',', dialect='excel')
    writer.writerow(header_row)
    
    # Construct row from user data
    for i, ID in enumerate(user_ids):
        if i > 100:
            break

        user = client.get_user(id=ID, user_fields=user_fields)
        json_user = user.json()
        row = []
        row.append(ID)
        try:
            data = json_user['data']
            
            # Get bio
            try:
                row.append(data['description'])
            except:
                row.append('')
            
            # Get location
            try:
                row.append(data['location'])
            except:
                row.append('')
                
            # Get name
            try:
                row.append(data['name'])
            except:
                row.append('')
                
            # Get user name
            try:
                row.append(data['username'])
            except:
                row.append('')
                
            # Get protected status
            try:
                row.append(data['protected'])
            except:
                row.append('')
                
            # Get verified status
            try:
                row.append(data['verified'])
            except:
                row.append('')
                
            # Get public metrics
            try:
                metrics = data['public_metrics']
                
                # Get followers count
                try:
                    row.append(metrics['followers_count'])
                except:
                    row.append('')
                    
                # Get friends count
                try:
                    row.append(metrics['following_count'])
                except:
                    row.append('')
                    
                # Get statuses count
                try:
                    row.append(metrics['tweet_count'])
                except:
                    row.append('')
                
            except:
                pass

        except:
            pass
        
        writer.writerow(row)
```
